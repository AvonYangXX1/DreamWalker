```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
jaccard = pd.read_csv('drive/MyDrive/DreamWalker/jaccard_df.csv', index_col=0)
```

```python
sns.clustermap(jaccard, xticklabels=False, yticklabels=False,
               cmap=sns.color_palette("vlag", as_cmap=True))
plt.savefig('jaccard.png')
plt.show()
```
